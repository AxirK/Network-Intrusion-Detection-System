```python
import pandas as pd
import pickle
```

```python
# features of NUSW-NB15
df_col = ['srcip', 'sport', 'dstip', 'dsport', 'proto', 'state', 'dur', 'sbytes', 'dbytes', 'sttl',\
         'dttl', 'sloss', 'dloss', 'service', 'Sload', 'Dload', 'Spkts', 'Dpkts', 'swin', 'dwin',\
         'stcpb', 'dtcpb', 'smeansz', 'dmeansz', 'trans_depth', 'res_bdy_len', 'Sjit', 'Djit', 'Stime', 'Ltime',\
         'Sintpkt', 'Dintpkt', 'tcprtt', 'synack', 'ackdat', 'is_sm_ips_ports', 'ct_state_ttl', 'ct_flw_http_mthd', 'is_ftp_login', 'ct_ftp_cmd',\
         'ct_srv_src', 'ct_srv_dst', 'ct_dst_ltm', 'ct_src_ ltm', 'ct_src_dport_ltm', 'ct_dst_sport_ltm', 'ct_dst_src_ltm', 'attack_cat', 'Label']
drop_col = ['srcip','sport','dstip','dsport','state',\
                'service', 'trans_depth', 'res_bdy_len', 'Stime', 'Ltime',\
                'is_sm_ips_ports', 'ct_state_ttl', 'ct_flw_http_mthd', 'is_ftp_login', 'ct_ftp_cmd',\
                'ct_srv_src', 'ct_srv_dst', 'ct_dst_ltm', 'ct_src_ ltm', 'ct_src_dport_ltm',\
                'ct_dst_sport_ltm', 'ct_dst_src_ltm', 'attack_cat']
```

```python
# each part of the dataset has 700001 rows * 49 cols
df1 = pd.read_csv("UNSW-NB15_1.csv", names=df_col)
df2 = pd.read_csv("UNSW-NB15_2.csv", names=df_col)
df3 = pd.read_csv("UNSW-NB15_3.csv", names=df_col)
```

```python
# drop unwanted cols
df = pd.concat([df1,df2,df3], ignore_index=True) # 2100003 rows * 49 cols
df = df.drop( drop_col, axis=1) # 2100003 rows * 26 cols
df.fillna(value=0)
```

```python
Y = df.pop('Label') # 2100003 rows * 1 cols
# onehot encode 'proto' of all three parts of data
# since df3 has no col 'esp' from onehot 'proto'
df = df.join(pd.get_dummies(df.proto))
X = df.drop('proto', axis=1) # 2100003 rows * 159 cols
```

```python
## Cloud Layer: Batch Learning using Random Forest
# train with part-1 data
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=100, max_depth=30, random_state=0)
RF.fit(X.iloc[:700001,:],Y.iloc[:700001])
```

```python
pickle.dump(RF, open('./cloudRF.pickle', 'wb'))
```

```python
# predict labels of part-2 data
ypred2 = RF.predict(X.iloc[700001:1400002,:])
```

```python
# performance of cloud layer model on part-2 data
from sklearn import metrics
print('accuracy: ', metrics.accuracy_score(Y.iloc[700001:1400002], ypred2))
print('recall: ', metrics.recall_score(Y.iloc[700001:1400002], ypred2))
print('precision: ', metrics.precision_score(Y.iloc[700001:1400002], ypred2))
print('f1: ', metrics.f1_score(Y.iloc[700001:1400002], ypred2))
print(metrics.classification_report(Y.iloc[700001:1400002], ypred2))
```

```python
from river import metrics
Accu = metrics.Accuracy()
Rec = metrics.Recall()
Prec = metrics.Precision()
F1 = metrics.F1()
for yt, yp in zip(Y.iloc[1400002:], ypred3):
    Accu = Accu.update(yt,yp)
    Rec = Rec.update(yt,yp)
    Prec = Prec.update(yt,yp)
    F1 = F1.update(yt,yp)
```

```python
# performance of proposed method without pre-training
print(Accu,Rec,Prec,F1)
```
